## Introduction

Software applications reuse a small number of primitives (instructions) over and over again in time to implement a given computation. The domain of computer architecture deals with designing those primitives used by software and implementing them so that they are efficient according to some relevant measure such as time, energy, or implementation size. This IIA project will give you an opportunity
to gain first-hand experience in modifying a popular computer architecture, the RV32I 32-bit integer variant of the RISC-V instruction set architecture (ISA).

The project will require you to first become familiar with the concepts of processor instructions, processor state, and processor microarchitecture. You will then explore how to implement basic hardware designs using the Verilog hardware description language and you will use this understanding to collaborate in a team to improve
the performance of your microarchitectural implementation of the RV32I ISA.

> RISC-V is pronounced “RISC five”.





## Learning objectives

The project will introduce you to the concepts and tools for designing and modeling microprocessors, implementing hardware designs in FPGAs and evaluating their hardware resource usage, timing performance, and power dissipation.

The aim of the project is to understand how a processor is designed using Verilog, get first-hand experience implementing a baseline processor design on an FPGA, and modifying the processor to improve its hardware resource usage, timing performance, and/or power dissipation.

In this project, you will:
* learn the key concepts of computer architecture, including the concept of an
instruction set architecture, processor state, and processor microarchitecture.
* learn how to use the Verilog hardware description language to design hardware.
* learn how to use open-source FPGA tools to synthesize and load Verilog hardware designs to the iCE40 FPGA.
* learn how to characterize the performance of a processor design.
* get experience optimising a processor design with respect to power consumption, timing performance, and resource usage.
* explore how to work effectively in a Team environment and identify appropriate roles for yourself

## Intended Learning Outcomes

At the end of this project, you should be able to:
* create new hardware designs in the Verilog hardware description language.
* synthesise Verilog hardware designs for FPGAs using open-source
FPGA tools.
* use your understanding of Verilog language and the related FPGA tools to modify a RISC-V processor design and evaluate it on an FPGA.
* propose and implement micro-architectural changes to a processor design to im-
prove the performance and resource usage of the processor design.

## iCEBreaker Hardware Specifications

These are copied from iCEBreaker README, but only crucial resources for this project are included.

* iCE40UP5K in QFN48 (SG48) package
  * iCE40 UltraPlus 5K [Datasheet PDF Download](http://www.latticesemi.com/-/media/LatticeSemi/Documents/DataSheets/iCE/iCE40-UltraPlus-Family-Data-Sheet.ashx)
  * **5280** Logic Cells (4-LUT + Carry + FF)
  * 128 KBit Dual-Port Block RAM
  * 1 MBit (128 KB) Single-Port RAM
  * PLL, Two SPI and two I2C hard IPs
  * Two internal oscillators (10 kHz and 48 MHz)
  * 8 DSPs (16x16 multiply + 32 bit accumulate)
  * 3x 24mA drive and 3x hard PWM IP

* QSPI-DDR-capable flash 128 MBit (16 MB)
  * The Winbond
    [W25Q128JVSIM](http://www.winbond.com/resource-files/w25q128jv_dtr%20revc%2003272018%20plus.pdf) is used.
  * Projects may access the flash at the highest flash access speed possible.

* [FT2232H](http://www.ftdichip.com/Support/Documents/DataSheets/ICs/DS_FT2232H.pdf) interface (microUSB plug)
  * programming using **[iceprog](https://github.com/cliffordwolf/icestorm/tree/master/iceprog)** or [Dimond Programmer](http://www.latticesemi.com/programmer)
  * 12 MHz XTAL oscillator (shared with FPGA)

* 39 I/O capable pins:
  * 4 pins for config (SDI, SDO, SCK, CSB)
    * Either loading config from on board FLASH chip or provided through the
      FTDI chip with direct SRAM config.
  * 2 extra GPIO pins for QSPI
    * Together with the config pins, allows storage of additional data in the FLASH chip, and high speed QSPI DDR access. For example [picosoc](https://github.com/cliffordwolf/picorv32/tree/master/picosoc) firmware.
  * 3 PINs for RGB LED (pin header)
  * 2 LEDs (one on output-only PLL pin)
  * 1 Clock pin (on PLL GBIN)
  * 1 UART Rx Pin via FTDI
  * 1 UART Tx Pin via FTDI
  * 1 Push Button
  * 16 PINs on dual PMOD
  * 8 PINs on single PMOD / snap-off section

* Break-off/Snap-off section (convertible to PMOD host / PMOD device)
  * 5 LEDs in similar arrangement to iCEstick
  * 3 Push Buttons

* Unpopulated parts shipped with the board
  * 3x Host PMOD (2x for dual PMOD port, 1x for snap-off section)
  * 1x Device PMOD (for other side of snap-off section)

## RISCV GNU toolchain (MacOS, Homebrew)

These are the instructions if you wish to install the toolchain on your own laptops.

Last tested: 16.11.2025 (Macbook Air M3, Tahoe 26.1) by Matthew Tang.

Add the homebrew tap from https://github.com/riscv-software-src/homebrew-riscv and install the tools.

    brew tap riscv-software-src/riscv
    brew install riscv-tools
    brew edit riscv-gnu-toolchain

To install 32-bit cross-compiler, update the variables ``args`` to use ``rv32i`` architecture (32-bit integer) and ``ilp32`` (soft floating points)
```
args = [
    "--prefix=#{prefix}",
    "--with-cmodel=medany",
    "--with-arch=rv32i",
    "--with-abi=ilp32"
  ]
```
Then build the toolchain from source

    brew reinstall --build-from-source riscv-gnu-toolchain
